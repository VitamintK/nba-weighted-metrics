```python
import time
def avg_year(team_id, weighted=True, prec=2):
    #to-do: rounding should be done at presentation, not at raw data level
    time.sleep(0.2)
    #print(team_id)
    t = team.TeamPlayers(team_id,per_mode="Totals").season_totals()
    t["birthdate"] = t.apply(lambda x: birthdate(x.PLAYER_ID),axis=1)
    t["days_old"] = pd.datetime.now() - t.birthdate
    t["years_old"] = t.days_old.dt.days/365 #kind of an estimate cause leap years exist
    if weighted:
        t["year_minutes"] = (t.MIN * t.years_old)
        return round(t.year_minutes.sum()/t.MIN.sum(),prec)
    else:
        return round(t["years_old"].mean(),prec)
    
```

```python
teams = team.TeamList().info()
current_teams = teams[teams["MAX_YEAR"] == '2016']
```

```python
current_teams['OLDNESS'] = current_teams.apply(lambda x: avg_year(x.TEAM_ID),axis=1)
```

```python
def color_quantile(s):
    q = s.rank(pct=True)
    scaled_q = 2*(q - 0.5)
    return ['background-color: rgb({0},255,{0})'.format(int(255 - 255*v)) if v> 0 else 'background-color: rgb(255,{0},{0})'.format(int(255 + 255*v)) for v in scaled_q]
```
